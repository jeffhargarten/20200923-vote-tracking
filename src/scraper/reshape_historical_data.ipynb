```python
out_obj = {'ts_statewide': []}
counties = list(weekly_2018.county)
```

```python
# counties
```

```python
def process_week(ballots_requested, ballots_accepted, week_name):
    print(ballots_requested, ballots_accepted, week_name)
    out_obj[county]['apps_submitted_latest'] = ballots_requested
    out_obj[county]['ballots_accepted_latest'] = ballots_accepted

    ts_record = {
        'weeks_away': week_name,
        'apps_submitted': ballots_requested,
        'ballots_accepted': ballots_accepted,
    }

    out_obj[county]['ts'].append(ts_record.copy())

    ts_statewide_date = [date for date in out_obj['ts_statewide'] if date['weeks_away'] == week_name]
    if len(ts_statewide_date) == 0:
        out_obj['ts_statewide'].append(ts_record)
    else:
        # ts_statewide_date[0]['weeks_away'] += weeks_away(version_obj['data_date'])
        ts_statewide_date[0]['apps_submitted'] += out_obj[county]['apps_submitted_latest']
        ts_statewide_date[0]['ballots_accepted'] += out_obj[county]['ballots_accepted_latest']
```

```python
for county in counties:
    out_obj[county] = {
        'ts': [],
        'apps_submitted_latest': 0,
        'ballots_accepted_latest': 0,
    }
    for week in range(1, 8):
        ballots_requested = int(weekly_2018[weekly_2018['county'] == county]['week{}_requested'.format(week)])
        ballots_accepted = int(weekly_2018[weekly_2018['county'] == county]['week{}_ab'.format(week)])
        
        process_week(ballots_requested, ballots_accepted, 7-week)
        
    # Add election day
    ballots_requested = int(weekly_2018[weekly_2018['county'] == county]['electionday_requested'])
    ballots_accepted = int(weekly_2018[weekly_2018['county'] == county]['electionday_ab'])
    process_week(ballots_requested, ballots_accepted, 'election_day')
```

```python
out_obj
```

```python
with open('json/county-list-2018.json', 'w') as json_2018:
    json_2018.write(json.dumps(out_obj))
    json_2018.close()
```

```python
fridays_2016 = [
    {'friday_date': datetime.datetime.fromisoformat('2016-09-23'), 'weeks_away' : 7},
    {'friday_date': datetime.datetime.fromisoformat('2016-09-30'), 'weeks_away' : 6},
    {'friday_date': datetime.datetime.fromisoformat('2016-10-07'), 'weeks_away' : 5},
    {'friday_date': datetime.datetime.fromisoformat('2016-10-14'), 'weeks_away' : 4},
    {'friday_date': datetime.datetime.fromisoformat('2016-10-21'), 'weeks_away' : 3},
    {'friday_date': datetime.datetime.fromisoformat('2016-10-28'), 'weeks_away' : 2},
    {'friday_date': datetime.datetime.fromisoformat('2016-11-04'), 'weeks_away' : 1},
    {'friday_date': datetime.datetime.fromisoformat('2016-11-07'), 'weeks_away' : 0},
    {'friday_date': datetime.datetime.fromisoformat('2016-11-08'), 'weeks_away': 'election_day'},
]
```

```python
voter_file_2016 = pd.read_csv(
    'import/voters_ab_ballots_Nov7_2016.csv',
    parse_dates=['DateAccepted'],
    dtype={'RecordId': object, 'VoterId': object, 'CountyCode': object}
)
voter_file_2016.columns
```

```python
voter_file_2016.head()
```

```python
voter_file_2016_filtered = voter_file_2016[voter_file_2016['DateAccepted'] > '2016-09-01']
```

```python
def tag_week(input_date):
    for week in reversed(fridays_2016):
        if input_date <= week['friday_date']:
            which_week = week['weeks_away']
    return which_week
```

```python
voter_file_2016_filtered['weeks_away'] = voter_file_2016_filtered.DateAccepted.apply(lambda x: tag_week(x))
voter_file_2016_filtered
```

```python
county_week_totals_2016 = voter_file_2016_filtered[
    ['RecordId', 'CountyName', 'weeks_away']
].groupby([
    'CountyName', 'weeks_away'
]).agg('count').reset_index().sort_values(['CountyName', 'weeks_away'])
county_week_totals_2016.rename(columns={'RecordId': 'count'}, inplace=True)
county_week_totals_2016
```

```python
county_week_totals_2016 = voter_file_2016_filtered[
    ['RecordId', 'CountyName', 'weeks_away']
].groupby([
    'CountyName', 'weeks_away'
]).agg('count').sort_values(['CountyName', 'weeks_away'], ascending=[True, False]).groupby(level=0).cumsum().reset_index()
county_week_totals_2016.rename(columns={'RecordId': 'running_count'}, inplace=True)
county_week_totals_2016
```

```python
statewide_weekly_totals_2016 = voter_file_2016_filtered[
    ['RecordId', 'weeks_away']
].groupby(['weeks_away']).agg('count').sort_values('weeks_away', ascending=False).cumsum().reset_index()
# statewide_weekly_totals_2016.rename(columns={'RecordId': 'running_count'}, inplace=True)
statewide_weekly_totals_2016
```
